```python
from sklearn import cross_validation, datasets, linear_model, metrics

data, target, coef = datasets.make_regression (n_features=2,n_informative =1 ,n_targets=1,noise=5,coef=True,random_state=2)
```

```python
print(data[:5],target[:5])
```

```python
pylab.scatter(list(map(lambda x:x[0],data)),target,color='r')
pylab.scatter(list(map(lambda x:x[1],data)),target,color='b')
```

```python
train_data,test_data,train_label,test_label = cross_validation.train_test_split(data,target,test_size=0.3)
```

```python
lineal_regression = linear_model.LinearRegression()
lineal_regression.fit(train_data,train_label)
prediction = lineal_regression.predict(test_data)
```

```python
print(test_label)
```

```python
print(prediction)
```

```python
metrics.mean_absolute_error(test_label,prediction)
```

```python
lineal_scoring=cross_validation.cross_val_score(lineal_regression,data,target,scoring=scorer,cv=100)
print('mean {}, std: {}'.format(lineal_scoring.mean(),lineal_scoring.std()))
```

```python
scorer=metrics.make_scorer(metrics.mean_absolute_error,greater_is_better=True)
```

```python
coef
```

```python
lineal_regression.coef_
```

### lasso (лучше если есть много избыточных признаков)

```python
lasso_regression = linear_model.Lasso(random_state=3)
lasso_regression.fit(train_data,train_label)
las_predict=lasso_regression.predict(test_data)
```

```python
print(las_predict)
```

```python
las_scor=cross_validation.cross_val_score(lasso_regression,data,target,scoring=scorer,cv=10)
print('mean {}, std: {}'.format(las_scor.mean(),las_scor.std()))
```

```python
print('y={:.2f}*x1 + {:.2f}*x2'.format(coef[0],coef[1]))
```

```python
print('y={:.2f}*x1 + {:.2f}*x2'.format(lasso_regression.coef_[0],lasso_regression.coef_[1]))
```
